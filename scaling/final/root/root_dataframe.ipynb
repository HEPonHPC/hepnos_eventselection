```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

pwd = os.getcwd()
```

```python
#https://stackoverflow.com/a/6402859
def get_sec(time_str):
    """Get seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

```python
nodes = np.array([8,16,32,64,128,256])
```

```python
times_root = pd.DataFrame(
    {
        "files" : [],
        #"server-nodes-fraction" : [],
        #"server-rpc-xstreams-per-MPI-rank" : [],
        "total_num_nodes" : [],
        #"MPI-rank-among-clients" : [],
        "expt_number" : [],
        #"pep_start_times" : [],
        #"pep_end_times" : [],
        "analysis_end_times" : [],
        "analysis_method" : [],
    }
)
```

```python
#1929-files

#8,16,32,64,128,256
es1929_analysis_end_times = [[],[],[],[],[],[]]

# 8 node run, 5_18_600136
_time_8_nodes = get_sec("16:46:25") - get_sec("16:39:55") 

_df_8_ranks = pd.DataFrame(
    {
        "analysis_end_times": [_time_8_nodes],
        "expt_number" : [1],
        "files" : [1929],
        "total_num_nodes" : [8],
        "analysis_method" : str("ROOT"),
    }
)
times_root = pd.concat([times_root, _df_8_ranks], ignore_index=True)


f = open('es1929/611672.output')
l = f.readlines()
f.close()

locs = []
for i in range(len(l)):
    if l[i].split()[1] == 'before':
        locs.append(i)    
for i in locs:
    _time = get_sec(l[i+1].split()[-3]) - get_sec(l[i].split()[-3])
    if l[i].split()[4] == '16':
        es1929_analysis_end_times[1].append(_time)
    elif l[i].split()[4] == '32':
        es1929_analysis_end_times[2].append(_time)
    elif l[i].split()[4] == '64':
        es1929_analysis_end_times[3].append(_time)
    elif l[i].split()[4] == '128':
        es1929_analysis_end_times[4].append(_time)
    elif l[i].split()[4] == '256':
        es1929_analysis_end_times[5].append(_time)
        

for i in range(5):
    _nodes = nodes[i+1]
    for j in range(5):
        _df = pd.DataFrame(
            {
                "analysis_end_times": [es1929_analysis_end_times[i+1][j]],
                "expt_number" : [j+1],
                "files" : [1929],
                "total_num_nodes" : [_nodes],
                "analysis_method" : str("ROOT"),
            }
        )
        # results with 256 nodes are not valid!
        if _nodes!=256:
            times_root = pd.concat([times_root, _df], ignore_index=True)
```
