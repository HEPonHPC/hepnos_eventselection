```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
hepnos_v1_1929_mean_times = np.load('hepnos/v1/means_1929.npy')
hepnos_v1_1929_errs_times = np.load('hepnos/v1/errs1929.npy')

hepnos_v1_3858_mean_times = np.load('hepnos/v1/means_3858.npy')
hepnos_v1_3858_errs_times = np.load('hepnos/v1/errs3858.npy')

hepnos_v1_7716_mean_times = np.load('hepnos/v1/means_7716.npy')
hepnos_v1_7716_errs_times = np.load('hepnos/v1/errs7716.npy')
```

```python
hepnos_v2_1929_mean_times = np.load('hepnos/v2/means_1929.npy')
hepnos_v2_1929_errs_times = np.load('hepnos/v2/errs1929.npy')

hepnos_v2_3858_mean_times = np.load('hepnos/v2/means_3858.npy')
hepnos_v2_3858_errs_times = np.load('hepnos/v2/errs3858.npy')

hepnos_v2_7716_mean_times = np.load('hepnos/v2/means_7716.npy')
hepnos_v2_7716_errs_times = np.load('hepnos/v2/errs7716.npy')
```

```python
root_1929_mean_times = np.load('root/root_es1929_mean.npy')
root_1929_errs_times = np.load('root/root_es1929_errs.npy')

root_3858_mean_times = np.load('root/root_es3858_mean.npy')
root_3858_errs_times = np.load('root/root_es3858_errs.npy')

root_7716_mean_times = np.load('root/root_es7716_mean.npy')
root_7716_errs_times = np.load('root/root_es7716_errs.npy')
```

```python
nodes = [8,16,32,64,128]

mpl.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 0.5
mpl.rcParams['lines.markersize'] = 6

f,ax = plt.subplots(1,3)
f.set_size_inches(32,10)
f.subplots_adjust(wspace=0)

_elinewidth = 0.75
_capsize = 3

ax[0].set_title('1929 files')
ax[0].errorbar(np.arange(5), hepnos_v1_1929_mean_times, yerr=hepnos_v1_1929_errs_times, fmt='b-.d', ecolor='b', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v1')
ax[0].plot(np.arange(5), (hepnos_v1_1929_mean_times[0]*nodes[0])/nodes, 'b--.', alpha=0.35)
ax[0].errorbar(np.arange(5), hepnos_v2_1929_mean_times, yerr=hepnos_v2_1929_errs_times, fmt='g-o', ecolor='g', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v2')
ax[0].plot(np.arange(5), (hepnos_v2_1929_mean_times[0]*nodes[0])/nodes, 'g--.', alpha=0.35)
ax[0].errorbar(np.arange(5), root_1929_mean_times, yerr=root_1929_errs_times, fmt='m-x', ecolor='m', elinewidth=_elinewidth, capsize=_capsize, label='ROOT')
ax[0].plot(np.arange(5), (root_1929_mean_times[0]*nodes[0])/nodes, 'm--.', alpha=0.35)

ax[1].set_title('3858 files')
ax[1].errorbar(np.arange(5), hepnos_v1_3858_mean_times, yerr=hepnos_v1_3858_errs_times, fmt='b-.d', ecolor='b', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v1')
ax[1].plot(np.arange(5), (hepnos_v1_3858_mean_times[0]*nodes[0])/nodes, 'b--.', alpha=0.35)
ax[1].errorbar(np.arange(5), hepnos_v2_3858_mean_times, yerr=hepnos_v2_1929_errs_times, fmt='g-o', ecolor='g', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v2')
ax[1].plot(np.arange(5), (hepnos_v2_3858_mean_times[0]*nodes[0])/nodes, 'g--.', alpha=0.35)
ax[1].errorbar(np.arange(5), root_3858_mean_times, yerr=root_3858_errs_times, fmt='m-x', ecolor='m', elinewidth=_elinewidth, capsize=_capsize, label='ROOT')
ax[1].plot(np.arange(5), (root_3858_mean_times[0]*nodes[0])/nodes, 'm--.', alpha=0.35)

ax[2].set_title('7716 files')
ax[2].errorbar(np.arange(5), hepnos_v1_7716_mean_times, yerr=hepnos_v1_7716_errs_times, fmt='b-.d', ecolor='b', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v1')
ax[2].plot(np.arange(5), (hepnos_v1_7716_mean_times[0]*nodes[0])/nodes, 'b--.', alpha=0.35)
ax[2].errorbar(np.arange(5), hepnos_v2_7716_mean_times, yerr=hepnos_v2_7716_errs_times, fmt='g-o', ecolor='g', elinewidth=_elinewidth, capsize=_capsize, label='HEPnOS-v2')
ax[2].plot(np.arange(5), (hepnos_v2_7716_mean_times[0]*nodes[0])/nodes, 'g--.', alpha=0.35)
ax[2].errorbar(np.arange(5), root_7716_mean_times, yerr=root_7716_errs_times, fmt='m-x', ecolor='m', elinewidth=_elinewidth, capsize=_capsize, label='ROOT')
ax[2].plot(np.arange(5), (root_7716_mean_times[0]*nodes[0])/nodes, 'm--.', alpha=0.35)

# Common legend for all plots!
ax[1].legend(bbox_to_anchor=(0.925, -0.125), ncol=3)

for _ax in ax:
    _ax.set_xticks(np.arange(5))
    _ax.set_xticklabels(nodes)
    _ax.set_xlabel('Nodes', labelpad=10)
    _ax.set_yscale('log', base=10)
    _ax.set_ylim(10,1500)
    _ax.xaxis.set_tick_params(which='major', size=10, width=2)
    _ax.xaxis.set_tick_params(which='minor', size=7, width=2)
    _ax.yaxis.set_tick_params(which='major', size=10, width=2)
    _ax.yaxis.set_tick_params(which='minor', size=7, width=2)
    if _ax==ax[0]:
        _ax.set_ylabel('PEP time in sec', labelpad=10)
    else:
        _ax.yaxis.set_ticklabels([])
        _ax.yaxis.set_ticks([])
        _ax.yaxis.set_ticks([], minor=True)
```
